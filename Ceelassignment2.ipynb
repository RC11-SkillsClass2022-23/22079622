```python
import math
def harmonic_number(n):
    H = 0
    for i in range(1, n+1):
        H += 1/i
    return H
x = int(input("Please enter one natural number: "))
print(harmonic_number(x))
```

The asymptotic time complexity of this function is O(n), because it performs a simple loop over n elements and adds 1/i to a sum. Each iteration of the loop takes constant time, so the overall time complexity is linear with respect to the input size n.

product of all positive integers less than or equal to n such as 

( n! = n × (n − 1)!

```python
import math
def factorial(n):
    if n == 0:
        return 1
    else:
        return int(n) * factorial(int(n) - 1)

x = int(input("Please enter one non-negative integer"))
factorial(x)
```

Bubble Sort is a comparison-based sorting algorithm. The algorithm works by comparing each pair of adjacent elements and swapping them if they are in the wrong order. The algorithm repeats this process until all elements are sorted in the right order.

The cost of the algorithm can be analyzed by counting the number of swaps performed during each iteration. On the first iteration, the largest element is sorted and placed at the end of the list, so the number of swaps performed is equal to the number of elements in the list minus one. On the second iteration, the second largest element is sorted and placed in the second to last position, so the number of swaps performed is equal to the number of elements minus two. The number of swaps performed decreases by one each iteration, until the number of swaps performed on the final iteration is equal to zero.

Therefore, the total number of swaps performed is equal to the sum of the first n-1 positive integers, which is equal to (n-1) * n / 2. The time complexity of the algorithm can be expressed as O(n^2), as each swap takes O(1) time, and the total number of swaps is proportional to n^2.

In conclusion, the asymptotic complexity of Bubble Sort is O(n^2), which makes it less efficient than other sorting algorithms, such as Quick Sort and Merge Sort, which have an asymptotic complexity of O(n log n).
