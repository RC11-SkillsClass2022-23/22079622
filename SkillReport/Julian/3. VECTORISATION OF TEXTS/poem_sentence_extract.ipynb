```python
import os
import re

folder_path = "C:/Users/user/Downloads/code/Istanbul_Books_txt"

all_sentences = ""  # Initialize an empty string to hold all the sentences

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):  # Consider only text files
        file_path = os.path.join(folder_path, filename)
        with open(file_path, "r", encoding="utf-8") as f:
            poem_text = f.read()

        # Split the poem into sentences using regular expressions
        sentences = re.split(r'[.!?]+', poem_text)

        # Combine all the sentences from the poem into the all_sentences string
        all_sentences += "\n".join(sentences)

# Print out the combined string of all sentences
print(all_sentences)
```

```python
import os
import re

folder_path = "C:/Users/user/Downloads/code/poem"

a_sentences = ""  # Initialize an empty string to hold all the sentences

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):  # Consider only text files
        file_path = os.path.join(folder_path, filename)
        with open(file_path, "r", encoding="utf-8") as f:
            poem_text = f.read()

        # Split the poem into sentences using regular expressions
        sentences = re.split(r'[.!?]+', poem_text)

        # Combine all the sentences from the poem into the all_sentences string
        a_sentences += "\n".join(sentences)

# Split the all_sentences string into a list of sentences
sentence_list = a_sentences.split("\n")

all_sentences = []
current_item = ""

for line in sentence_list:
    if line.strip() == "":
        if current_item != "":
            all_sentences.append(current_item)
            current_item = ""
    else:
        current_item += line

if current_item != "":
    all_sentences.append(current_item)

print(all_sentences)
```

```python
import os
import re

folder_path = "C:/Users/user/Downloads/code/poem"

sentence_poem_index = []  # Initialize an empty list to hold the poem index for each sentence

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):  # Consider only text files
        file_path = os.path.join(folder_path, filename)
        with open(file_path, "r", encoding="utf-8") as f:
            poem_text = f.read()

        # Split the poem into sentences using regular expressions
        sentences = re.split(r'[.!?]+', poem_text)

        # Add the sentences from the poem to the all_sentences list
        for sentence in sentences:
            if sentence.strip() != "":
                sentence_poem_index.append(filename)

# Print number of sentences and unique poem files
print(f'Number of sentences: {len(all_sentences)}')
print(f'Number of unique poem files: {len(set(sentence_poem_index))}')

# Print first 10 sentences and corresponding poem files
for i in range(10):
    print(f'Sentence {i+1}: {all_sentences[i]}')
    print(f'Poem file: {sentence_poem_index[i]}')
```

```python
import pprint
stoplist = set('for a of the and to in'.split(' '))

texts = [[word.replace(".","").replace(",","") for word in document.lower().split()] 
         for document in all_sentences]

texts = [[word for word in text if (word not in stoplist and len(word)>2)] 
         for text in texts]

to_delete = []
for i in range(len(texts)):
    t = texts[i]
    test = [w for w in t if w.isalpha()]
    if len(test) < 20:
        to_delete.append(i)
    else:
        texts[i] = test

for i in sorted(to_delete, reverse = True):
    del texts[i]
    del all_sentences[i]
    
# Count word frequencies
from collections import defaultdict
frequency = defaultdict(int)
for text in texts:
    for token in text:
        frequency[token] += 1

# Only keep words that appear more than once
processed_corpus = [[token for token in text if frequency[token] > 1] for text in texts]
pprint.pprint(processed_corpus[0])
```

```python
from gensim import corpora

dictionary = corpora.Dictionary(processed_corpus)
print(dictionary)
```

```python
pprint.pprint(dictionary.token2id)
```

```python
bow_corpus = [dictionary.doc2bow(text) for text in processed_corpus]
```

```python
from gensim import models

tfidf = models.TfidfModel(bow_corpus)
```

```python
from gensim.similarities import Similarity
from gensim.test.utils import get_tmpfile
```

```python
index_tmpfile = get_tmpfile("index")

index = Similarity(index_tmpfile, bow_corpus, num_features = len(dictionary))
```

```python
query_document = "Cash machine".lower().split()
query_bow = dictionary.doc2bow(query_document)
```

```python
sims = index[query_bow]
```

```python
for document_number, score in sorted(enumerate(sims), key=lambda x: x[1], reverse=True)[:250]:
    print(document_number, score)
```

```python
# sort the scores in descending order and get the top 10 scores
top_scores = sorted(enumerate(sims), key=lambda x: x[1], reverse=True)[:250]
```

```python
# create a list of the top 10 paragraphs
top_sentences = [all_sentences[idx] for idx, score in top_scores]

# create a list of the corresponding books for the top 10 paragraphs
top_poems = [sentence_poem_index[idx] for idx, score in top_scores]

# combine the top paragraphs and their corresponding books into a single string
output = ""
for i in range(len(top_sentences)):
    output += f"Poem {top_poems[i]} - Sentence {i+1}: {top_sentences[i]}\n"

# write the output to a file named "top_paragraphs.txt"
with open("Tea kiosk.txt", "w", encoding="utf-8") as f:
    f.write(output)

# print the output to the console
print(output)
```
